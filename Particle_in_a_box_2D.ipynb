```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook

# Constants
l = 1
n_x = 1
n_y = 1

# Define the spatial grid
x = np.linspace(0, l, 100)
y = np.linspace(0, l, 100)
x, y = np.meshgrid(x, y)

# Calculate the wave function
psi = np.sqrt(2 / l) * np.sin(n_x * np.pi * x / l) * np.sqrt(2 / l) * np.sin(n_y * np.pi * y / l)

# Plot the wave function
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, psi, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('$\psi(x, y)$')
plt.title(f'Wave function for $n_x = {n_x}$, $n_y = {n_y}$')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import ipywidgets as widgets
from ipywidgets import interactive

# Constants
l = 1

def plot_2D_wavefunction(n_x, n_y):
    # Define the spatial grid
    x = np.linspace(0, l, 100)
    y = np.linspace(0, l, 100)
    x, y = np.meshgrid(x, y)

    # Calculate the wave function
    psi = np.sqrt(2 / l) * np.sin(n_x * np.pi * x / l) * np.sqrt(2 / l) * np.sin(n_y * np.pi * y / l)

    # Plot the wave function
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(x, y, psi, cmap='viridis', alpha=0.6)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('$\psi(x, y)$')
    plt.title(f'Wave function for $n_x = {n_x}$, $n_y = {n_y}$')
    plt.show()

# Create sliders
n_x_slider = widgets.IntSlider(min=1, max=5, step=1, value=1, description='$n_x$')
n_y_slider = widgets.IntSlider(min=1, max=5, step=1, value=1, description='$n_y$')

# Interactive plot
interactive_plot = interactive(plot_2D_wavefunction, n_x=n_x_slider, n_y=n_y_slider)
output = interactive_plot.children[-1]
output.layout.height = '500px'
interactive_plot
```
