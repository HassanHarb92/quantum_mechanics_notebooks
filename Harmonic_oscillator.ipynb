```python
import numpy as np
import matplotlib.pyplot as plt
#from scipy.constants import hbar

# Constants
m = 1          # Mass of the particle
omega = 1      # Angular frequency
x = np.linspace(-3.5, 3.5, 1000) # Range of x values
n_levels = 5   # Number of energy levels to display
hbar=1

# Quantum harmonic oscillator potential function
def potential(x):
    return 0.5 * m * omega**2 * x**2

# Energy level function
def energy_level(n):
    return (n + 0.5) * hbar * omega
# Plotting the potential
plt.plot(x, potential(x), label="Potential")
plt.title("Potential and Energy Levels for a Quantum Harmonic Oscillator")
plt.xlabel("Position (x)")
plt.ylabel("Energy")

# Plotting the energy levels
for n in range(n_levels):
    energy = energy_level(n)
    plt.hlines(energy, x[0], x[-1], color="grey", linestyle="--", label=f"n = {n}")

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#from scipy.constants import hbar
from math import pi, sqrt

# Constants
m = 1          # Mass of the particle
omega = 1      # Angular frequency
x = np.linspace(-5, 5, 1000) # Range of x values
hbar = 1
# Quantum harmonic oscillator potential function
def potential(x):
    return 0.5 * m * omega**2 * x**2

# Energy level function
def energy_level(n):
    return (n + 0.5) * hbar * omega

# Ground state wavefunction
def wavefunction_0(x):
    return (m * omega / (pi * hbar))**(1/4) * np.exp(-m * omega * x**2 / (2 * hbar))

# Plotting the potential
plt.plot(x, potential(x), label="Potential")
plt.title("Potential and Ground State Wavefunction for a Quantum Harmonic Oscillator")
plt.xlabel("Position (x)")
plt.ylabel("Energy / Amplitude")

# Plotting the energy level
energy_0 = energy_level(0)
plt.hlines(energy_0, x[0], x[-1], color="green", linestyle="--", label="n = 0")

# Plotting the ground state wavefunction
plt.plot(x, wavefunction_0(x) + energy_0, label="Wavefunction for n = 0") # Shifted by energy_0

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite

# Constants
m = 1          # Mass of the particle
omega = 1      # Angular frequency
hbar = 1       # Set to 1 for simplicity
x = np.linspace(-5, 5, 1000) # Range of x values

# Quantum harmonic oscillator potential function
def potential(x):
    return 0.5 * m * omega**2 * x**2

# Energy level function
def energy_level(n):
    return (n + 0.5) * hbar * omega

# Wavefunction
def wavefunction(n, x):
    prefactor = (m * omega / (np.pi * hbar))**(1/4) / np.sqrt(2**n * np.math.factorial(n))
    hermite_poly = hermite(n)
    return prefactor * hermite_poly((m * omega / hbar)**0.5 * x) * np.exp(-m * omega * x**2 / (2 * hbar))

# Setting the figure size
plt.figure(figsize=(12, 8))

# Plotting the potential
plt.plot(x, potential(x), label="Potential")
plt.title("Potential and Wavefunctions for a Quantum Harmonic Oscillator")
plt.xlabel("Position (x)")
plt.ylabel("Energy / Amplitude")

# Plotting the energy levels and wavefunctions for n = 0 to 4
for n in range(5):
    energy_n = energy_level(n)
    plt.hlines(energy_n, x[0], x[-1], color="grey", linestyle="--")
    plt.plot(x, wavefunction(n, x) + energy_n) # Shifted by energy_n

plt.ylim(0)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite
import ipywidgets as widgets
from IPython.display import display

# Constants
m = 1          # Mass of the particle
omega = 1      # Angular frequency
hbar = 1       # Set to 1 for simplicity
x = np.linspace(-5, 5, 1000) # Range of x values

# Quantum harmonic oscillator potential function
def potential(x):
    return 0.5 * m * omega**2 * x**2

# Energy level function
def energy_level(n):
    return (n + 0.5) * hbar * omega

# Wavefunction
def wavefunction(n, x):
    prefactor = (m * omega / (np.pi * hbar))**(1/4) / np.sqrt(2**n * np.math.factorial(n))
    hermite_poly = hermite(n)
    return prefactor * hermite_poly((m * omega / hbar)**0.5 * x) * np.exp(-m * omega * x**2 / (2 * hbar))

# Function to update the plot based on the selected value of n
def update_plot(n):
    plt.figure(figsize=(12, 8))
    plt.plot(x, potential(x), label="Potential")
    plt.title("Potential and Wavefunctions for a Quantum Harmonic Oscillator")
    plt.xlabel("Position (x)")
    plt.ylabel("Energy / Amplitude")
    
    # Plotting all energy levels as dashed lines
    for level in range(11): 
        energy_n = energy_level(level)
        plt.hlines(energy_n, x[0], x[-1], color="grey", linestyle="--")
    
    # Plotting the selected wavefunction
    plt.plot(x, wavefunction(n, x) + energy_level(n), color="green") 

    plt.ylim(0)
    plt.legend()
    plt.grid(True)
    plt.show()

# Creating a slider for selecting the value of n
n_slider = widgets.IntSlider(value=0, min=0, max=10, step=1, description='n:')
interactive_plot = widgets.interactive(update_plot, n=n_slider)
display(interactive_plot)
```
